```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Mani/Downloads/Iris.csv")
```

```python
df
```

```python
from sklearn.decomposition import PCA
```

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
x = df.loc[:, features].values
y = df.loc[:,['Species']].values
```

```python
x
```

```python
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x)
principalComponents
```

```python
df1 = pd.DataFrame(data = principalComponents, columns = ['1','2','3'])
```

```python
df1
```

```python
df2 = pd.concat([df1, df[['Species']]], axis = 1)
```

```python
df2
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('1', fontsize = 15)
ax.set_ylabel('2', fontsize = 15)
ax.set_zlabel('3', fontsize = 15)
ax.set_title('3 component PCA', fontsize = 20)
Species = ['Iris-setosa','Iris-versicolor','Iris-virginica']
colors = ['r', 'g', 'b']
for Species, color in zip(Species,colors):
    indicesToKeep = df2['Species'] == Species
    ax.scatter(df2.loc[indicesToKeep,'1'],df2.loc[indicesToKeep,'2'],df2.loc[indicesToKeep,'3'],c = color,s = 50)
ax.legend(labels = ['Iris-Setosa','Iris-versicolor','Iris-virginica'])
ax.grid()
plt.show()
```
